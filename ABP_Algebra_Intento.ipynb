Preparación del Entorno 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "C:/Users/solvc/Downloads/archive.zip"
Primer_Proyecto = pd.read_csv(url, encoding="latin-1") # Usamos encoding='latin1' para evitar errores al leer archivos con caracteres especiales (como tildes o eñes) 
print(Primer_Proyecto.head())  # Vista inicial
```

Conociendo nuestro Dataset

```python
print("Tamaño del DataFrame:", Primer_Proyecto.info)
print("\nÍndice del DataFrame:", Primer_Proyecto.shape)
print("\nTipos de datos de las columnas:\n", Primer_Proyecto.columns)
print("Tamaño del DataFrame:", Primer_Proyecto.size)
print("\nÍndice del DataFrame:", Primer_Proyecto.index)
print("\nTipos de datos de las columnas:\n", Primer_Proyecto.dtypes)
```

```python
print("Estadísticas descriptivas para columnas numéricas:")
display(Primer_Proyecto.describe())

print("\nEstadísticas descriptivas para columnas de tipo 'object':")
display(Primer_Proyecto.describe(include='object'))
```

Valores Nulos y Duplicados 

```python
# Verificar valores nulos por columna
print("Valores nulos por columna:")
print(Primer_Proyecto.isnull().sum())

# Verificar filas duplicadas
print("\nCantidad de filas duplicadas:")
print(Primer_Proyecto.duplicated().sum())
```
